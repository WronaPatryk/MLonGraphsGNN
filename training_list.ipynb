# Training list of audio files ( .wav format) creation

```python
import pandas as pd, numpy as np
```

```python
# chosen audio files classes
used_commands = ['yes', 'no', 'up', 'down', 
                 'left', 'right', 'on', 'off', 
                 'stop', 'go']
```

```python
# vl - validation list from kaggle, used to load seach audio file
vl = list(pd.read_csv("validation_list.txt", header = None).iloc[:,0])
# tl - testing list from kaggle, used to load seach audio file
tl = list(pd.read_csv("testing_list.txt",header = None).iloc[:,0])
```

```python

checkList = []

for v in vl:
    checkList.append(v)
for t in tl:
    checkList.append(t)
```

```python
len(checkList)
```

```python
from os import listdir
from os.path import isfile, join

PATH_FOLDER = "audio/"

training_audiofiles = []

file_counter = 0

for x in used_commands:
    
    PATH_FOLDER2 = PATH_FOLDER + x + "/"
    audiofiles_x = [f for f in listdir(PATH_FOLDER2) if isfile(join(PATH_FOLDER2, f))]

    file_counter+=len(audiofiles_x)
    
    # determining training set:
    
    for audiofile_x in audiofiles_x:
        
        aux = x + "/" + audiofile_x
        
        if aux in checkList:
#             print("already in test/valid")
            continue
        else:
            training_audiofiles.append(aux)
    
    


```

```python
len(checkList)
```

```python
len(training_audiofiles)
```

```python
file_counter
```

```python
# train + test/val == all files

len(training_audiofiles) + len(checkList) == file_counter
```

```python

pd.DataFrame(training_audiofiles).to_csv("training_list.txt", header = None, index=None)
    
```
