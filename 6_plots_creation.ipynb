```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# M dependence on test accuracy

```python
data = pd.read_csv("metrics/m_dependence_on_test.csv")

data2 = pd.read_csv("metrics/m_dependence_on_test-spectr3.csv")
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(data.iloc[:,0], data.iloc[:,1], marker='.')
ax.plot(data2.iloc[:,0], data2.iloc[:,1], marker='.')


ax.legend( ['spectrograms - no. 2','spectrograms - no. 3'])

ax.set_title("Test Accuracy vs M in nearest neighbors, (spectrogram 2 & 3) data")
ax.set_ylabel("Test Accuracy")
ax.set_xlabel("M - Number of nearest neighbors")


plt.show()
```
